```python
import pandas as pd

reddit = pd.read_csv("RedditNews.csv", engine="python", sep=",", error_bad_lines=False)
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk.corpus import stopwords
import re
import numpy as np
```

```python
lemmatizer = WordNetLemmatizer()

def preprocess(sentence):
    sentence=str(sentence)
    sentence = sentence.lower()
    sentence=sentence.replace('{html}',"") 
    cleanr = re.compile('<.*?>')
    cleantext = re.sub(cleanr, '', sentence)
    rem_url=re.sub(r'http\S+', '',cleantext)
    tokenizer = RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(rem_url)  
    lemma_words=[lemmatizer.lemmatize(w) for w in tokens]
    return " ".join(lemma_words)


reddit['cleanText']=reddit['Title'].map(lambda s:preprocess(s))
```

```python
reddit['Date'] =  pd.to_datetime(reddit['Date'])
reddit.head()
```

```python
foo = lambda a: " ".join(a) 
```

```python
pd.options.display.max_colwidth = 100
```

```python
new_reddit = reddit.groupby(by='Date').agg(foo).reset_index()
new_reddit['cleanText'].head()
```

```python
new_reddit.sort_values(by=['Date'], ascending = True)
```

```python
djia = pd.read_csv("DJIA.csv", engine="python", sep=",", error_bad_lines=False)
```

```python
djia['Date'] =  pd.to_datetime(djia['Date'], format='%Y-%m-%d')
```

```python
djia = djia.sort_values(by=['Date'], ascending = True)
```

```python
returns = (djia.Close - djia.Close.shift(1))/djia.Close.shift(1)
djia['returns'] = returns
djia['volatility'] = np.log(djia['High'] - djia['Low'])
```

```python
final_ascending = djia.sort_values(by=['Date'], ascending = True)
final_ascending.head()
```

```python
merge=pd.merge(djia,new_reddit, how='inner')
merge.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(ngram_range = (1, 2), max_features = 7000)
vectorizer.fit(merge['cleanText'])
print(vectorizer.vocabulary_)
```

```python
vec = vectorizer.transform(merge['cleanText'])
```

```python
#the word_count array
dayvectors = vec.toarray()
```

```python
dayvectors.shape
```

```python
#dictionary of words and their ID's
vocab = vectorizer.vocabulary_
vocab
```

# Machine Learning Part

## Select features (columns)

```python
X = dayvectors
y1 = merge['returns'].values.reshape(-1, 1)
y2 = merge['volatility'].values.reshape(-1, 1)
```

## Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale the data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

X_train, X_test, y1_train, y1_test, y2_train, y2_test = train_test_split(X, y1, y2, random_state=1)
```

```python
X_scaler = MinMaxScaler().fit(X_train)
y1_scaler = MinMaxScaler().fit(y1_train)
y2_scaler = MinMaxScaler().fit(y2_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y1_train_scaled = y1_scaler.transform(y1_train)
y1_test_scaled = y1_scaler.transform(y1_test)
y2_train_scaled = y2_scaler.transform(y2_train)
y2_test_scaled = y2_scaler.transform(y2_test)
```

## Train the model

```python
# LASSO model (return)
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error

lasso = Lasso().fit(X_train_scaled, y1_train_scaled)
predictions = lasso.predict(X_test_scaled)

MSE = mean_squared_error(y1_test_scaled, predictions)
r2 = lasso.score(X_test_scaled, y1_test_scaled)

print(f"MSE: {MSE}, R2: {r2}")
```

```python
# LASSO model (volatility)

lasso = Lasso(alpha=0.5).fit(X_train_scaled, y2_train_scaled)

predictions = lasso.predict(X_test_scaled)

MSE = mean_squared_error(y2_test_scaled, predictions)
r2 = lasso.score(X_test_scaled, y2_test_scaled)

print(f"MSE: {MSE}, R2: {r2}")
```

```python
#Lasso with cross validation (return)
from sklearn.linear_model import LassoCV

lasso_return = LassoCV(cv = 5, n_jobs = -1).fit(X_train_scaled, y1_train_scaled)

predictions = lasso_return.predict(X_test_scaled)

MSE = mean_squared_error(y1_test_scaled, predictions)
r2 = lasso_return.score(X_test_scaled, y1_test_scaled)

print(f"MSE: {MSE}, R2: {r2}")
```

```python
#Lasso with cross validation (volatility)

lasso_volatility = LassoCV(cv = 5, n_jobs = -1).fit(X_train_scaled, y2_train_scaled)

predictions = lasso_volatility.predict(X_test_scaled)

MSE = mean_squared_error(y2_test_scaled, predictions)
r2 = lasso_volatility.score(X_test_scaled, y2_test_scaled)
### END SOLUTION

print(f"MSE: {MSE}, R2: {r2}")
```

```python
# retrieve the top 50 predictors for both return and volatility model
indices_return = np.argsort(abs(lasso_return.coef_))[-50:]
indices_volatility = np.argsort(abs(lasso_volatility.coef_))[-50:]

top_50_return = np.array(list(vocab.keys()))[indices_return]
top_50_volatility = np.array(list(vocab.keys()))[indices_volatility]
```
